#### Correlations

```python
#Correlations among numeric fields
numeric_df = df.select_dtypes(include=['number'])
correlation = numeric_df.corr()
correlation
```

```python
#Heatmap of the above
sns.heatmap(correlation, annot=True)
plt.show()
```

#### Grouping

```python
#Average of all numeric fields based on country
numeric_df_and_country = numeric_df.copy()
numeric_df_and_country['Country'] = df['Country']
avg_by_country_df = numeric_df_and_country.groupby('Country').mean()
avg_by_country_df
```

```python
#Average of all numeric fields based on country, ordered by least expensive tuition
avg_bot = avg_by_country_df.sort_values(by="Tuition_USD").head(10)
avg_bot
```

```python
#Average of all numeric fields based on country, ordered by most expensive tuition
avg_top = avg_by_country_df.sort_values(by="Tuition_USD", ascending=False).head(10)
avg_top
```

```python
#All records where country is 'USA'
USA_df = df[df['Country'].str.contains('USA')]
USA_df
```

```python
#Unique count of each field where country is 'USA'
USA_df.nunique()
```

```python
USA_df.columns
```

#### Visualizations

```python
#Bar plot for country averages based on lowest tuition
bar_avg_bot = avg_bot.plot(kind='bar', subplots=True, figsize=(20,10), title='')
for subplot in bar_avg_bot:
    subplot.set_title('')
```

```python
#Bar plot for country averages based on highest tuition
bar_avg_top = avg_top.plot(kind='bar', subplots=True, figsize=(20,10), title='')
for subplot in bar_avg_top:
    subplot.set_title('')
```

```python
#Boxplot of average tuition and exchange rate based on country
avg_by_country_df[['Tuition_USD', 'Exchange_Rate']].boxplot(figsize=(20,20))
```

```python
#Boxplot of average duration in years, living cost index, rent, and visa fee based on country
avg_by_country_df[['Duration_Years',
        'Living_Cost_Index', 'Rent_USD', 'Visa_Fee_USD',
       'Insurance_USD']].boxplot(figsize=(20,20))
```
